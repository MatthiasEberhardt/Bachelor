# Space-Eye Ship Dataset

Here I describe the Spaceeye Ship Dataset.
It consists of 872 non-ship geotiffs and 872 ship geotiffs.
Those were downloaded from planet.com by intersection with AIS files from marineCadastre.gov.
https://marinecadastre.gov/ais/ & https://www.planet.com/

The dataset contains all ships between Miami and Palm Beach in 2017 (just those that were caught on planet).

The ship and nonship .csv files contain all the information provided by marincadastre.gov and in addition the "newindex" columns which refer to the index under which the (non)ship can be found in the (non)ship folder. It also contains a column named "envelope" which is the coordinates of the rectangle used to download/request the data from planet. The "url" column is a backup for redownload and does not have any meaning to the user, same goes for the "nonship" column.

All other metadata is described here.

!

To read in the images there are multiple options.
Each folder contains multiple files:

- ..._AnalyticsMS_clip.jpg - a randomly cut out RGB image as jpg
- ..._AnalyticsMS_clip.npy - a randomly cut out 4 channel image as npy
- ..._AnalyticsMS_clip.tif - the full geotif that was downloaded from planet
- ..._AnalyticsMS_DN_udm_clip.tif - the non usable pixels inside the geotif 
- ..._AnalyticsMS_metadata_clip.xml - the metadata for the geotif
- ..._manifest.json
- ..._metadata.json

The jpg and npy files are nice for a first impression but for training the full geotiffs provide more freedom (random crops, additional channel,...)

The metadata, manifest and unusable pixels probably won't be needed.

There are multiple options to read and work on a geotiff file. Those include:
- PIL if the file has 'tiff' as an ending (so renaming would be necessary)
- rasterio which gives access to a lot of additional information

There are more options for sure and for the ML application speed is our highest concern. 

Another "problem" with the tiff files is there shape - They are not rectangles. This is due to the curvature of the earth. A coordinate rectangle is no rectangle in pixels. So every tiff is shaped slightly differently. In the downloading process all ships were placed in the middle of the coordinate rectangles so they can safeley be thought of as inside the middle pixels of the geotiffs and the geotiffs are big enough to just cut some rectangle from the full tiff.
