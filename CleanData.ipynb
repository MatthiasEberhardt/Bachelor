```python
def get_4ch_tif_files(path):
    file_paths=[]
    for root,directories,files in os.walk(path):
        #print(root)
        file_paths+=[Path(join(root,f)) for f in files if ".tif" in f and "udm" not in f]
    return L(file_paths)

def get_data_y(item):
    return Path(item).parent.parent.name
```

```python
def get_max_min(files):
    max_,min_=0,10000
    for f in files:
        #print(get_data_y(f))
        #ship=(get_data_y(f)=="ships")
        with rasterio.open(f) as data:
            d=data.read()
            loc_max,loc_min=np.amax(d),np.amin(d)
            if loc_max>max_:
                max_=loc_max
            if loc_min<min_:
                min_=loc_min
    return max_,min_
```
